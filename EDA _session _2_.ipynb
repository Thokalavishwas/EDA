**task_1**

- using dtypes

- whenever you want to create any list from series 
- first convertb the series into a dictionary
-  when you convert it becomes a kry:valuenpair
- iterate though dictionary
- if that value==objecttype,then keep that corresponding key in one list
- else keep that variables in other list

```python
data_types=visa_df.dtypes
dict(data_types)
```

```python
cat_list=[]
num_list=[]
for i,j in dict(data_types).items():
    if j=='object':
        cat_list.append(i)
    else:
        num_list.append(i)
    
    
```

```python
cat_list=[i for i,j in dict(data_types).items() if j=='object']
num_list=[i for i,j in dict(data_types).items() if j!='object']
cat_list,num_list
```

```python
for i,j in dict(data_types).items():
    print(i,j)
```
